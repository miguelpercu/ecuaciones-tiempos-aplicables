```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import time
import pandas as pd

# Iniciar el cronmetro
start_time = time.time()

# Constantes cosmolgicas (valores de ejemplo basados en parmetros tpicos)
H0 = 67.4  # Constante de Hubble en km/s/Mpc
H0 = H0 * 3.24078e-20  # Convertir a s^-1 (1 km/s/Mpc = 3.24078e-20 s^-1)
Omega_m0 = 0.27  # Densidad de materia oscura hoy
Omega_DE0 = 0.68  # Densidad de energa oscura hoy
rho_c0 = 3 * H0**2 / (8 * np.pi * 6.67430e-11)  # Densidad crtica hoy (en kg/m^3)
M0 = Omega_m0 * rho_c0  # Densidad de materia oscura hoy (rho_DM0)
N0 = Omega_DE0 * rho_c0  # Densidad de energa oscura hoy (rho_DE0)

# Imprimir M0 y N0
print(f"M0 (rho_DM0) = {M0:.2e} kg/m^3")
print(f"N0 (rho_DE0) = {N0:.2e} kg/m^3")

# Rango de redshift z
z = np.linspace(0, 1089, 1000)

# Modelo cosmolgico: ecuaciones para rho_DM y rho_DE
# rho_DM(z) = rho_DM0 * (1+z)^3
rho_DM = M0 * (1 + z)**3

# rho_DE(z) = rho_DE0 * (1+z)^(3*(1+w)), con w = -1.03 (por ejemplo)
w0 = -1.03
rho_DE = N0 * (1 + z)**(3 * (1 + w0))

# Densidad total
rho_total = rho_DM + rho_DE

# Calcular H(z)
def H(z, H0, Omega_m0, Omega_DE0, w0):
    return H0 * np.sqrt(Omega_m0 * (1 + z)**3 + Omega_DE0 * (1 + z)**(3 * (1 + w0)))

H_z = H(z, H0, Omega_m0, Omega_DE0, w0)

# Calcular t_applied usando un bucle con quad
t_applied = np.zeros_like(z)
for i in range(len(z)):
    integrand = lambda zp: 1 / ((1 + zp) * H(zp, H0, Omega_m0, Omega_DE0, w0))
    t, _ = quad(integrand, z[i], np.inf, epsabs=1e-6, epsrel=1e-6)
    t_applied[i] = t

# Imprimir algunos valores de t_applied
print("Valores de t_applied (s):")
print(f"z={z[0]:.1f}: t_applied={t_applied[0]:.1e} s")
print(f"z={z[-1]:.1f}: t_applied={t_applied[-1]:.1e} s")

# Graficar rho_DM, rho_DE, rho_total y H(z)
plt.figure(figsize=(10, 12))

# Grfico de rho_DM
plt.subplot(4, 1, 1)
plt.plot(t_applied, rho_DM, 'b-', label=r'$\rho_{DM}$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Tiempo Aplicado (s)')
plt.ylabel(r'$\rho_{DM}$ (kg/m^3)')
plt.legend()
plt.grid(True)

# Grfico de rho_DE
plt.subplot(4, 1, 2)
plt.plot(t_applied, rho_DE, 'r-', label=r'$\rho_{DE}$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Tiempo Aplicado (s)')
plt.ylabel(r'$\rho_{DE}$ (kg/m^3)')
plt.legend()
plt.grid(True)

# Grfico de rho_total
plt.subplot(4, 1, 3)
plt.plot(t_applied, rho_total, 'g-', label=r'$\rho_{total}$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Tiempo Aplicado (s)')
plt.ylabel(r'$\rho_{total}$ (kg/m^3)')
plt.legend()
plt.grid(True)

# Grfico de H(z)
plt.subplot(4, 1, 4)
plt.plot(z, H_z, 'm-', label=r'$H(z)$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Redshift z')
plt.ylabel(r'$H(z)$ (s^-1)')
plt.legend()
plt.grid(True)

# Ttulo general para los grficos
plt.suptitle("Evolucin de las Densidades y H(z)")

# Ajustar el diseo para evitar solapamiento
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Mostrar el grfico
plt.show()

# Guardar los datos en un archivo CSV en el escritorio
datos = pd.DataFrame({
    'Tiempo_Aplicado_s': t_applied,
    'Redshift_z': z,
    'rho_DM_kg_m3': rho_DM,
    'rho_DE_kg_m3': rho_DE,
    'rho_total_kg_m3': rho_total,
    'H_z_s^-1': H_z
})
datos.to_csv(r'C:\Users\los percu\Desktop\datos_densidades_evolucion.csv', index=False)

# Calcular y mostrar el tiempo de ejecucin
end_time = time.time()
execution_time = end_time - start_time
print(f"Tiempo de ejecucin: {execution_time:.3f} segundos")
```
